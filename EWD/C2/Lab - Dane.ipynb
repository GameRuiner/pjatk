## Brakujące wartości
Brakujące dane w pandas zapisywane są jako `NaN`.
 - Wykrywanie brakujących:`df.info()`, `df.isna()`.
 - Usuwanie (wiersze lub kolumny): `df.dropna(axis=0)`, `df.dropna(axis=1)`.
 - Zastępowanie podaną wartością: `df.fillna()`.

```python
housing_na_rows = housing.isna().any(axis=1)
housing[housing_na_rows]
```

```python
census = pd.read_csv("censusData-with3errors", na_values=['?'], skipinitialspace=True)
census.info()
```

```python
census_na = census.isna().sum(axis = 1)
census_na[census_na.iloc[:] == census_na.max()]
```

```python
census_na.value_counts()
```

```python
census_cols = census.isna().sum(axis=0) / census.count(axis=0)
cols = census_cols[census_cols.iloc[:] > 0.03]
cols
```

```python
for col in cols.index:
  census[col] = census[col].fillna(census[col].mode()[0])
```

```python
census_cols = census.isna().sum(axis=0) / census.count(axis=0)
census_cols
```
