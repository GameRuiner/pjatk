```python
import matplotlib.pyplot as plt

df_dimensions = df[['x dimension', 'y dimension', 'z dimension']].dropna(axis=0)
tick_labels = df_dimensions.columns
plt.boxplot(df_dimensions, labels=tick_labels, vert=False)
plt.show();
# looks like no outliers there
```

```python
q = df.price.quantile(0.97)
df_2 = df[df.price < q]
```

```python
plt.boxplot(df_2.price.dropna(axis=0), labels=["Price"]);
# outliers cleaned from price
```

```python
df_2[df_2.carat.isna()]
```

```python
df_2[['x dimension', 'y dimension', 'z dimension']].corr()
# as features strongly correlated we can use it to fullfil empty values
```

```python
df_2
```

```python
df_3 = df_2.copy()
df_3['x dimension'].fillna((df_3['y dimension'] + df_3['z dimension']) / 2, inplace=True)
```

```python
df_3['x dimension']
```

```python
df_4 = df_3.copy()
df_4['y dimension'].fillna((df_4['x dimension'] + df_4['z dimension']) / 2, inplace=True)
```

```python
df_5 = df_4.copy()
df_5['z dimension'].fillna((df_5['x dimension'] + df_5['y dimension']) / 2, inplace=True)
```

```python
df_5['z dimension'].info()
```

```python
df_6 = df_5.copy()
# create average dimension for predicting carat
df_6['dimension'] = (df_6['x dimension'] + df_6['y dimension'] + df_6['z dimension']) / 3 
```

```python
df_3 = df_6.drop(['x dimension', 'y dimension', 'z dimension'], axis=1)
df_3
```

```python
df_3[['dimension', 'carat']].corr()
# as correlation is quite high we can estimate carat based on dimension
```

```python
df_pred = df_3.dropna(subset=['carat'])
df_pred.describe()
```

```python
df_train = df_pred.dimension
df_output = df_pred.carat
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(df_train.values.reshape(-1, 1), df_output)
```

```python
test_data = df_3[df_3['carat'].isnull()].dimension
res = lr.predict(test_data.values.reshape(-1, 1))
res
```

```python
df_4 = df_3.copy()
df_4.loc[df['carat'].isna(), 'carat'] = res
df_4
```

```python
df_4.describe(include='all')
```

```python
df_4[['depth', 'table', 'dimension', 'carat']].corr()
```

```python
df_4[['depth', 'table', 'price', 'carat']].corr()
# and depth has significant correlation with carat similar to price lets drop it
```

```python
df_5 = df_4.drop('depth', axis=1)
df_5.describe(include='all')
# we can leave table for future consideration as it it correlating with price more than with carat
# lets fill it with median
```

```python
df_6 = df_5.copy() 
df_6['table'].fillna(df_6['table'].mean(), inplace=True)
df_6.describe(include='all')
```

```python
# dimension we don't need anymore 
df_6 = df_6.drop('dimension', axis=1) 
df_6.head()
```

```python
# Create a scatter plot with different colors based on the 'color' feature
unique_colors = df_6['color'].unique()
color_map = plt.get_cmap('viridis', len(unique_colors))
for color, group in df_6.groupby('color'):
    color_value = color_map(list(unique_colors).index(color))
    plt.scatter(group['carat'], group['price'], label=color, color=color_value, alpha=0.5);
plt.legend()
plt.show()
```
